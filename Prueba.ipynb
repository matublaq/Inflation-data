```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo
df = pd.DataFrame({'X': [1, 2, 3, 4, 5], 'Y': [2, 4, 6, 8, 10], 'Z': [1, 3, 5, 7, 9]})

# Obtener los datos de los elementos a graficar
x = df['X']
y = df['Y']
z = df['Z']

# Crear el gráfico con colores diferentes para cada conjunto de datos
plt.plot(x, y, color='blue', label='Elemento 1')
plt.plot(x, z, color='red', label='Elemento 2')

# Agregar leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```
