```python
import numpy as np
import pandas as pd
import matplotlib as plt

from googletrans import Translator
from datetime import datetime, date, timedelta
```

## <font color='white'>Columns and rows</font>

```python
CPI.rename(columns={"Unnamed: 0":"Country"}, inplace=True)
CPI
```

#### <font color='lightblue'>Traduction</font>

```python
translator = Translator()
countrys = []

for i in CPI["Country"]:
    countrys.append(translator.translate(i, dest='en').text) #scr="es" is automatic. Google translator automatically detects the language
print(countrys)


CPI["Country"] = countrys
CPI
```

```python
print(CPI.select_dtypes(exclude=['float']).columns)

chinaCPI = CPI[CPI['Country'] == 'China']
euCPI = CPI[CPI['Country'] == 'European Union']
japanCPI = CPI[CPI['Country'] == 'Japan']
usaCPI = CPI[CPI['Country'] == 'USA']
worldCPI = CPI[CPI['Country'] == 'World']
```
