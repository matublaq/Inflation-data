### Acomodando la columna fecha
### Fixing date column

```python
Fecha = list(BA1Y["Fecha"])
print(Fecha)

for i in range(len(Fecha)):
    Fecha[i] = Fecha[i].replace(".", " ")
    Fecha[i] = datetime.strptime(Fecha[i], "%d %m %Y").date()
    
    #day, month, year = Fecha[i].split(".")
    #Fecha[i] = datetime(int(year), int(month), int(day)).date()

Fecha = pd.to_datetime(Fecha, format="%Y %m %d")

Fecha = Fecha.sort_values(ascending=True)

print(Fecha)
print(type(Fecha[0]))
```

```python
BA1Y["Fecha"] = Fecha
type(BA1Y["Fecha"])
```

### Areglando las columnas de apertura, maximo, minimo y variacion
### Fixing open, close, max, min and variation columns

##### First variation

```python
new_var = list(BA1Y["% var."])
print(len(new_var))
print(new_var)
```

```python
inter = 0
print(new_var)
for i in range(0, len(new_var)):
    new_var[i] = new_var[i].split("%")
    if(inter == 0):
        print(new_var)
    
    new_var[i] = new_var[i][0]
    if(inter == 0):
        inter = 1
        print(new_var)

    new_var[i] = round(float(new_var[i])/100, 4)
print("\n")
print(new_var)
```

```python
BA1Y = BA1Y.assign(Variation=new_var)
BA1Y["Variation"]
BA1Y
```

##### now: Open, min, max

```python
################################## Apertua ##################################
Apertura = pd.array(BA1Y['Apertura'])
print(type(Apertura[0]))
print(Apertura)
for i in range(0, len(Apertura)):
    Apertura[i] = Apertura[i].replace(",", ".")
Apertura = Apertura.astype(float)

print(type(Apertura[0]))
print(Apertura)
################################## Minimo ##################################
Minimo = pd.array(BA1Y['Mínimo'])
print(type(Minimo[0]))
print(Minimo)
for i in range(0, len(Minimo)):
    Minimo[i] = Minimo[i].replace(",", ".")
Minimo = Minimo.astype(float)

print(type(Minimo[0]))
print(Minimo)
################################## Maximo ##################################
Maximo = pd.array(BA1Y['Máximo'])
print(type(Maximo[0]))
print(Maximo)
for i in range(0, len(Maximo)):
    Maximo[i] = Maximo[i].replace(",", ".")
Maximo = Maximo.astype(float)

print(type(Maximo[0]))
print(Maximo)
```

##### Fix done

### Hay datos de la tabla mal puestos, vamos a corregirlos

```python
BA1Y[BA1Y["Max"]>10]

```

```python
#Las diferencias encontradas son las mismas?
if (BA1Y[BA1Y["Open"]>1].equals(BA1Y[BA1Y['Close']>1])):
    print(True)
else:
    print(False)
```

```python
#Cuales son esas diferencias?
comparacion = pd.merge(BA1Y[BA1Y["Open"]>1], BA1Y[BA1Y["Close"]>1], how='outer', indicator=True)
diferencias = comparacion[comparacion['_merge'] != 'both']
diferencias
```

```python
#Separamos los errores
Open_left = diferencias[diferencias['_merge'] == 'left_only']['Open']
Close_right = diferencias[diferencias['_merge'] == 'right_only']['Close']
print(Open_left, "\n", Close_right)

#Misma cantidad de errores?
print(len(Open_left)==len(Close_right))

#Acomodamos los errores
for i in range(len(Open_left)):
    Open_left.iloc[i] = Open_left.iloc[i]/10
    Close_right.iloc[i] = Close_right.iloc[i]/10

diferencias.loc[diferencias['_merge'] == 'left_only', 'Open'] = Open_left
diferencias.loc[diferencias['_merge'] == 'right_only', 'Close'] = Close_right
diferencias
```
